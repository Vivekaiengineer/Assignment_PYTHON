```python
# Function to find the GCD of two numbers using a for loop
def find_gcd(num1, num2):
    # Determine the smaller number
    smaller = min(num1, num2)
    gcd = 1  # Initialize GCD as 1 (smallest possible divisor)

    # Loop from 1 to the smaller number
    for i in range(1, smaller + 1):
        # Check if i divides both numbers
        if num1 % i == 0 and num2 % i == 0:
            gcd = i  # Update gcd if condition is satisfied

    return gcd

# Input: Ask the user to enter two numbers
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

# Calculate and display the GCD
gcd = find_gcd(num1, num2)
print(f"The GCD of {num1} and {num2} is: {gcd}")
```

```python
# Function to check if a string is a palindrome
def is_palindrome(input_string):
    # Convert the string to lowercase to make it case-insensitive
    input_string = input_string.lower()

    # Calculate the length of the string
    length = len(input_string)

    # Check if the string reads the same forwards and backwards
    for i in range(length // 2):  # Only iterate up to the middle of the string
        if input_string[i] != input_string[length - 1 - i]:
            return False  # Not a palindrome if characters don't match

    return True  # Palindrome if all characters match

# Input: Get the string from the user
user_input = input("Enter a string: ")

# Check and display the result
if is_palindrome(user_input):
    print(f"The string '{user_input}' is a palindrome.")
else:
    print(f"The string '{user_input}' is not a palindrome.")
```

```python
# Python program to calculate the square root of a number using a for loop

def calculate_square_root(number):
    if number < 0:
        return "Square root of a negative number is not real."

    # Iterate through possible square roots
    for i in range(0, number + 1):
        # Check if i * i equals the number
        if i * i == number:
            return f"The square root of {number} is {i}."

        # If i * i exceeds the number, return the approximate square root
        if i * i > number:
            return f"The square root of {number} is approximately {i - 1}."

# Input: User provides the number
num = int(input("Enter a positive integer: "))

# Output: Call the function and display the result
result = calculate_square_root(num)
print(result)
```

```python
# Function to find the Least Common Multiple (LCM) of two numbers
def find_lcm(num1, num2):
    # Determine the greater of the two numbers
    if num1 > num2:
        greater = num1
    else:
        greater = num2

    # Start from the greater number and find the smallest number divisible by both
    for i in range(greater, (num1 * num2) + 1):
        if i % num1 == 0 and i % num2 == 0:
            return i

# Input: Ask the user to enter two positive integers
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

# Calculate and display the LCM
lcm = find_lcm(num1, num2)
print(f"The LCM of {num1} and {num2} is: {lcm}")
```

```python
# Function to determine the type of a triangle based on its sides
def triangle_type(side1, side2, side3):
    # Check if it's a valid triangle
    if side1 + side2 <= side3 or side1 + side3 <= side2 or side2 + side3 <= side1:
        return "Not a valid triangle"

    # Determine the type of triangle
    if side1 == side2 == side3:
        return "Equilateral"
    elif side1 == side2 or side1 == side3 or side2 == side3:
        return "Isosceles"
    else:
        return "Scalene"

# Input: Ask the user to enter the sides of the triangle
try:
    side1 = float(input("Enter the first side of the triangle: "))
    side2 = float(input("Enter the second side of the triangle: "))
    side3 = float(input("Enter the third side of the triangle: "))

    # Get the type of triangle
    result = triangle_type(side1, side2, side3)

    # Output the result
    print(f"The triangle is {result}")

except ValueError:
    print("Please enter valid numbers for the sides.")
```

```python
import cmath  # Import the complex math module to handle complex numbers

# Function to find and determine the roots of a quadratic equation
def quadratic_roots(a, b, c):
    # Calculate the discriminant
    discriminant = cmath.sqrt(b**2 - 4*a*c)

    # Calculate the roots using the quadratic formula
    root1 = (-b + discriminant) / (2 * a)
    root2 = (-b - discriminant) / (2 * a)

    return root1, root2

# Get input from the user
print("Quadratic equation in the form Ax^2 + Bx + C = 0")
a = float(input("Enter the coefficient A: "))
b = float(input("Enter the coefficient B: "))
c = float(input("Enter the coefficient C: "))

# Ensure the coefficient 'a' is not zero
if a == 0:
    print("Coefficient 'A' cannot be zero. It's not a quadratic equation.")
else:
    root1, root2 = quadratic_roots(a, b, c)
    print(f"The roots of the quadratic equation {a}x^2 + {b}x + {c} = 0 are:")
    print(f"Root 1: {root1}")
    print(f"Root 2: {root2}")
```

```python
# Recursive function to find all permutations of a string
def find_permutations(s, prefix=""):
    # Base case: If the string is empty, print the prefix as a permutation
    if len(s) == 0:
        print(prefix)
    else:
        # Recursive case: Loop through each character in the string
        for i in range(len(s)):
            # Extract the character at index i
            char = s[i]
            # Remaining string after removing the current character
            remaining = s[:i] + s[i+1:]
            # Recursive call with updated prefix and remaining string
            find_permutations(remaining, prefix + char)

# Input: Ask the user to enter a string
input_string = input("Enter a string: ")

# Find and print all permutations of the input string
print(f"Permutations of '{input_string}':")
find_permutations(input_string)
```

```python
a.Local Variables:
A local variable is a variable defined inside a function or a block of code. It is only accessible within the function or the block in which it is declared. Once the function finishes executing, the local variable is destroyed and can no longer be accessed.
```

```python
def my_function():
    local_var = 10  # This is a local variable
    print(local_var)  # It can be accessed within the function

my_function()  # This will print 10
# print(local_var)  # This would result in an error because local_var is not defined outside the function
```

```python
b.Global Variables:
A global variable is defined outside any function or block and can be accessed by any part of the program, including functions. However, if a global variable is modified inside a function, you must declare it as global inside the function to avoid creating a local variable with the same name.
```

```python
global_var = 20  # This is a global variable

def my_function():
    print(global_var)  # Global variable can be accessed inside the function

my_function()  # This will print 20
print(global_var)  # This will also print 20 outside the function
```

# Lambda Functions and Higher-Order Functions:

```python
1.What are lambda functions in Python, and when are they typically used?
```

```python
Lambda functions in Python are small, anonymous functions that can have any number of arguments but only one expression. They are used to create functions on the fly without needing to formally define them with a name using the def keyword. Lambda functions are useful for situations where you need a simple function for a short period and don't want to formally define a full function.
```

```python
# Lambda function to add two numbers
add = lambda a, b: a + b
print("Sum of 5 and 3:", add(5, 3))

# Lambda function used with map to square each number in the list
numbers = [1, 2, 3, 4, 5]
squares = list(map(lambda x: x**2, numbers))
print("Squares of numbers:", squares)

# Lambda function used with filter to select even numbers
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print("Even numbers:", even_numbers)

# Lambda function used for sorting a list of tuples based on the second element
pairs = [(1, 2), (3, 1), (5, 6), (7, 4)]
pairs_sorted = sorted(pairs, key=lambda x: x[1])
print("Sorted pairs by second element:", pairs_sorted)
```

```python
3.Explain the concept of higher-order functions in Python, and provide an example?
```

```python
A higher-order function in Python is a function that either:

Takes one or more functions as arguments, or
Returns a function as its result.
```

```python
Some common examples of higher-order functions in Python are:

map()
filter()
reduce()
Functions like sorted() that take a custom key function.
```

```python
# Higher-order function that takes a function and a list as arguments
def apply_function_to_list(func, numbers):
    result = []
    for number in numbers:
        result.append(func(number))
    return result

# Function to double the value
def double(x):
    return x * 2

# Function to square the value
def square(x):
    return x * x

# Test the higher-order function
numbers = [1, 2, 3, 4, 5]

# Apply 'double' function
doubled_numbers = apply_function_to_list(double, numbers)
print("Doubled Numbers:", doubled_numbers)

# Apply 'square' function
squared_numbers = apply_function_to_list(square, numbers)
print("Squared Numbers:", squared_numbers)
```

# Function Documentation and Best Practices:

```python
1.Explain the purpose of docstrings in Python functions and how to write them?
```

```python
In Python, docstrings are used to document a function, class, or module. They provide a convenient way to explain the functionality of the code, describe input parameters, and specify the return values.
```

```python
Docstrings are essential for:

Code Readability: They make the code easier to understand for other developers (or even yourself) who may read it in the future.
Automatic Documentation: Tools like help() in Python can use docstrings to display the function's documentation, which is helpful for quick references.
Maintaining Code Quality: A clear and concise docstring improves the maintainability of the codebase by explaining the logic and purpose of each function, class, or module.
Standardized Format: Docstrings follow a specific format (PEP 257), which allows developers to consistently describe the functions behavior.
```

```python
# Function to calculate the factorial of a number
def factorial(n):
    """
    Calculate the factorial of a given number.

    The factorial of a non-negative integer n is the product of all positive integers less than or equal to n.
    The factorial of 0 is 1.

    Parameters:
    n (int): The number for which the factorial is to be calculated. Must be a non-negative integer.

    Returns:
    int: The factorial of the number n.
    
    Example:
    >>> factorial(5)
    120
    """
    
    # Checking if n is a negative number
    if n < 0:
        raise ValueError("Input must be a non-negative integer")
    
    # Base case: factorial of 0 is 1
    if n == 0:
        return 1
    
    # Recursive case: n! = n * (n-1)!
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

# Calling the function with an example
print(factorial(5))
```

```python
2.Describe some best practices for naming functions and variables in Python, including naming conventions and guidelines?
```

```python
///1. Use Descriptive Names
Variables: Variable names should be descriptive enough to give an understanding of the stored value.

Good Example: age, user_name, total_price
Bad Example: a, b, x, y (These don't convey any meaningful information)
Functions: Function names should clearly indicate the action or purpose of the function.

Good Example: calculate_area, get_user_input, send_email
Bad Example: foo, bar (These are placeholders and dont describe the function)
2. Use Snake Case for Variable and Function Names
Snake Case: Python follows the snake_case naming style where words are separated by underscores (_) and all characters are lowercase. This is preferred for variable and function names.
Good Example: calculate_total_price, get_user_data, user_input
Bad Example: calculateTotalPrice, getUserData (CamelCase style is typically used for classes in Python, not variables or functions)
3. Use CamelCase for Class Names
Class Names: Use the CamelCase naming convention for class names, where the first letter of each word is capitalized.
Good Example: UserProfile, EmployeeDetails, InvoiceManager
Bad Example: user_profile, employee_details (Snake_case is not recommended for class names)
4. Avoid Using Reserved Keywords
Python has a list of reserved keywords that cannot be used as variable or function names. Using these will result in errors.
Good Example: user_input, response
Bad Example: def, class, return, import (These are Python's reserved keywords)
5. Keep Names Short but Meaningful
While names should be descriptive, they should also be concise and not overly long. A good balance is important.
Good Example: user_data, file_path, total_price
Bad Example: the_information_about_the_users_of_the_system_for_data_storage (Too long and cumbersome)
6. Use Constants in All Uppercase Letters
Constants: Variables that are meant to be constants (values that do not change) should be written in all uppercase letters with underscores separating words.
Good Example: MAX_USERS, PI, API_KEY
Bad Example: maxUsers, pi, api_key (These are not standard for constants)
7. Function Names Should Be Action-Oriented
Function Names: A function should be named based on the action it performs, typically a verb followed by a noun or an object.
Good Example: fetch_data, send_message, calculate_area
Bad Example: data_fetch, message_send (These do not follow the verb-noun convention)
8. Use Meaningful and Unambiguous Names
Avoid ambiguous names or names that could be confusing to the reader.
Good Example: calculate_discount, get_user_details
Bad Example: do_stuff, temp
9. Use Plural for Collections
List/Array Names: When naming variables that represent collections, like lists or arrays, use plural forms to indicate that they contain multiple items.
Good Example: users, products, items
Bad Example: user, product, item (These would imply a single entity)
10. Use a Consistent Naming Scheme
In projects with multiple developers or for long-term maintainability, it's important to use a consistent naming scheme. This ensures that the code is easily understandable by all developers involved.
Consistency in naming conventions will prevent confusion and help others understand the code faster.
```
